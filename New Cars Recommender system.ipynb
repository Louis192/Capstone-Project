```python
import pandas as pd
Cars=pd.read_csv("../CAPSTONE-PROJECT/Cars.csv")
df=pd.read_csv("../CAPSTONE-PROJECT/df.csv")
```

```python
# set index as Car_model to prepare to normalize nutrition values
df = df.set_index('Car_Id')
df
```

```python
df1=df.drop(columns='Price')
```

```python
# import required packages
import pandas as pd
from scipy.spatial.distance import euclidean
from sklearn.preprocessing import normalize
from keras.preprocessing import image
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
# normalized Car features by columns
df_normalized = pd.DataFrame(normalize(df1, axis=0))
df_normalized.columns = df1.columns
df_normalized.index = df1.index
df_normalized.head()
```

```python
# show Car_models, Car Features and image of selected car_Models
def selected_Car(Car_Id):
    # setting the image path of the inputted recipe id
    image_path = "../CAPSTONE-PROJECT/Car-Images/{}.jpg"
    image_path = image_path.format(Car_Id)
    
    # load the image and turn the image into an array
    img = image.load_img(image_path)
    img = image.img_to_array(img, dtype='int')
    
    # show the image using the image array using plot
    fig, ax = plt.subplots(1,1)
    ax.imshow(img)
    ax.axis('off')
    
    # set index of recipe dataframe to recipe id to find the recipe name
    Car_df = Cars.set_index('Car_Id')
    
    # using inputted Car_Model, find the recipe name and output as the title for the plot
    x = "{}  {}".format(Car_Id, Car_df.at[Car_Id,'Car_Model'])
    ax.set_title(x)
```

```python
"""
Car Recommender based on different distance calculation approaches

df_normalized: normalized Car Features data
distance_method: distance calculation approach: euclidean
Car_Model: find similar Car_Models based on the selected Car_Model 
N: Top N recipe(s)

Return:
1) Car data of selected recipe and Top N recommendation, 
"""

def Car_recommender(distance_method, Car_Id, N):
    # create dataframe used to store distances between Car Features
    allFeatures = pd.DataFrame(df_normalized.index)
    
    # remove rows where index is equal to the inputted Car_Model
    allFeatures = allFeatures[allFeatures.Car_Id!= Car_Id]
    
    # add a distance column that states the inputted features distance with every other features
    allFeatures["distance"] = allFeatures["Car_Id"].apply(lambda x: distance_method(df_normalized.loc[Car_Id], df_normalized.loc[x]))
    
    # sort the allFeatures by distance and take N closes number of rows to put in the TopNRecommendation as the recommendations
    TopNRecommendation = allFeatures.sort_values(["distance"]).head(N)
    
    print(TopNRecommendation)
    
    # create dataframe with index as Car_Model to get car names
    Cars_df_normalized = Cars.set_index('Car_Id')
    
    # create list of recommended Car_Models with the inputted Car_Model for comparison
    Car_Id = [Car_Id]
    
    # list to store Car_Model and Car names
    Cars_list = []
    
    # list to store image paths of recommended recipes
    image_list = []
    
   
    # default path of images
    image_path =  "../CAPSTONE-PROJECT/Car-Images/{}.jpg"
    
    # for each cars in TopNRecommendation, input to defined lists
    for CarId in TopNRecommendation.Car_Id:
        Car_Id.append(CarId)   # list of Car Model of selected Car and recommended Car(s)
        Cars_list.append("{}  {}".format(CarId, Cars_df_normalized.at[CarId,'Car_Model']))
        image_list.append(image_path.format(CarId))
    
    #list to store image array
    image_array = []
    
    # for each image path in the image_list, turn it into an image array and store it in the image_array list
    for imagepath in image_list:
        img = image.load_img(imagepath)
        img = image.img_to_array(img, dtype='int')
        image_array.append(img)
    
    # draw each recommended image
    fig = plt.figure(figsize=(20,20))
    gs1 = gridspec.GridSpec(1, N)
    axs = []
    for x in range(N):
        axs.append(fig.add_subplot(gs1[x]))
        axs[-1].imshow(image_array[x])
    [axi.set_axis_off() for axi in axs]
    for axi, x in zip(axs, Cars_list):
        axi.set_title(x)
    
    # return dataframe with the inputted Cars and the recommended Car's normalized features values
    return df_normalized.loc[Car_Id, :], df_normalized.loc[Car_Id, :]
```

```python
selected_Car(36)
```
